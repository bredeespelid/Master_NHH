```python
from youtube_transcript_api import YouTubeTranscriptApi
import json

def get_captions(video_ids, language='no'):
    all_captions = {}
    
    for video_id in video_ids:
        try:
            transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[language])
            
            # Format the transcript data
            formatted_transcript = [
                {
                    'text': entry['text'],
                    'start': entry['start'],
                    'duration': entry['duration']
                }
                for entry in transcript
            ]
            
            all_captions[video_id] = formatted_transcript
        except Exception as e:
            all_captions[video_id] = f"Error: {str(e)}"
    
    return all_captions

# Example usage
video_ids = ['Eanity6gg7k&t=942s']  # Replace with your desired video IDs
captions = get_captions(video_ids)

# Convert to JSON
json_output = json.dumps(captions, ensure_ascii=False, indent=2)

# Print JSON output
print(json_output)

# Optionally, save to a file
with open('captions.json', 'w', encoding='utf-8') as f:
    f.write(json_output)
```

```python
from googleapiclient.discovery import build
from youtube_transcript_api import YouTubeTranscriptApi
import json

DEVELOPER_KEY = 'AIzaSyBzuIHNjwKVSaYemBbogf0x8yHAA_Np0yo'
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'

class Youtube:
    def __init__(self, playlist_id, max_results=50):
        self.playlist_id = playlist_id
        self.max_results = max_results
        self.youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=DEVELOPER_KEY)

    def get_playlist_items(self):
        results = self.youtube.playlistItems().list(
            part='snippet',
            playlistId=self.playlist_id,
            maxResults=self.max_results
        ).execute()

        videos = []
        for item in results['items']:
            video_id = item['snippet']['resourceId']['videoId']
            title = item['snippet']['title']
            url = f'https://www.youtube.com/watch?v={video_id}'
            videos.append({'title': title, 'url': url, 'video_id': video_id})

        return videos

    def get_captions(self, videos, language='no'):
        all_captions = {}
        
        for video in videos:
            video_id = video['video_id']
            try:
                transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[language])
                
                # Format the transcript data
                formatted_transcript = [
                    {
                        'text': entry['text'],
                        'start': entry['start'],
                        'duration': entry['duration']
                    }
                    for entry in transcript
                ]
                
                all_captions[video_id] = {
                    'title': video['title'],
                    'url': video['url'],
                    'captions': formatted_transcript
                }
            except Exception as e:
                all_captions[video_id] = {
                    'title': video['title'],
                    'url': video['url'],
                    'error': f"Error: {str(e)}"
                }
        
        return all_captions

    def main(self):
        videos = self.get_playlist_items()
        captions = self.get_captions(videos)
        
        # Convert to JSON
        json_output = json.dumps(captions, ensure_ascii=False, indent=2)
        
        # Print JSON output
        print(json_output)
        
        # Save to a file
        with open('playlist_captions.json', 'w', encoding='utf-8') as f:
            f.write(json_output)

# Spillelistens ID (hentet fra URL-en)
playlist_id = 'PLEj7GYvM31TkurScEQwyaafW5woWR6jC7'

# Opprett et Youtube-objekt og kjør hovedfunksjonen
yt = Youtube(playlist_id)
yt.main()
```
