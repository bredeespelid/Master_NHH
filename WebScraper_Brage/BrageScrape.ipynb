```python
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# Import google-colab-selenium and start the Chrome driver
import google_colab_selenium as gs
driver = gs.Chrome()

# URL to start from
base_url = "https://openaccess.nhh.no/nhh-xmlui/browse?rpp=20&offset=0&etal=-1&sort_by=2&type=type&value=Master+thesis&order=DESC"
driver.get(base_url)

# Initialize a list to store data
data = []

while True:
    # Get page source and parse with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find all articles on the page
    articles = soup.find_all("div", class_="artifact-description")

    for article in articles:
        # Extract title
        title_element = article.find("h4", class_="artifact-title").find("a")
        title = title_element.text.strip() if title_element else "Ikke funnet"

        # Extract abstract
        abstract_element = article.find("div", class_="artifact-abstract")
        abstract = abstract_element.text.strip() if abstract_element else "Ikke funnet"

        # Extract year
        date_element = article.find("span", class_="date")
        year = date_element.text.strip() if date_element else "Ikke funnet"

        # Break loop if year is 2020
        if year == "2015":
            print("Årstall 2015 funnet. Avslutter.")
            driver.quit()
            df = pd.DataFrame(data)
            df.to_excel("nhh_theses.xlsx", index=False, encoding="utf-8-sig")
            display(df.head())
            exit()

        # Append data to list
        data.append({"Tittel": title, "Abstrakt": abstract, "Årstall": year})

        # Print data for the current article
        print(f"Tittel: {title}\nAbstrakt: {abstract}\nÅrstall: {year}\n")

    # Find the next page button and navigate, if exists
    try:
        next_page = driver.find_element(By.CSS_SELECTOR, "li.next.pull-right a.next-page-link")
        next_page.click()
        time.sleep(2)  # Wait for the page to load
    except Exception:
        # Exit the loop if next page is not found
        break

# Close the driver
driver.quit()

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Save as CSV (optional)
df.to_excel("nhh_theses.xlsx", index=False)

# Display the DataFrame
df.head()
```
